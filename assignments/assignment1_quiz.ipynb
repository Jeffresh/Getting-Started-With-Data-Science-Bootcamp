### Question 4: Execute the given set of code to do some data cleaning and then answer the further questions:
Which of the following is the correct barplot for 'Education' column?
- The last one.

```python
# clean up column names that contain whitespace
df = marketing_data.copy()
df.columns = df.columns.str.replace(' ', '')
 
# transform Income column to a numerical
df['Income'] = df['Income'].str.replace('$', '')
df['Income'] = df['Income'].str.replace(',', '').astype('float')
```

```python
plt.bar(x=df['Education'].value_counts().index, height=df['Education'].value_counts().values)
```

```python
print('categories ', df['Education'].value_counts().index)
print('values ', df['Education'].value_counts().values)
print(df['Education'].value_counts())
```
